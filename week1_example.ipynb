```python
# Python Visualization Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Matplotlib example

# Set random seed for reproducibility
np.random.seed(0)

mu = 50      # mean
sigma = 2    # standard deviation
n = 100      # sample size

# Generate data from a normal distribution
x = np.random.normal(mu, sigma, size=n)

# Simple histogram
fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.hist(x, bins=20, density=True,
         histtype='stepfilled',
         facecolor='g',
         alpha=0.75)
ax1.set_title('Histogram of x')
ax1.set_xlabel('Value')
ax1.set_ylabel('Density')
plt.show()
```

```python
# Create a histogram with custom (unequally spaced) bin edges

bins = [44, 46, 48, 50, 52, 54, 56]

fig, ax2 = plt.subplots(figsize=(8, 4))
ax2.hist(x, bins=bins, density=True,
         histtype='bar',
         rwidth=0.8)
ax2.set_title('Histogram of x with custom bins')
ax2.set_xlabel('Value')
ax2.set_ylabel('Density')
plt.show()
```
